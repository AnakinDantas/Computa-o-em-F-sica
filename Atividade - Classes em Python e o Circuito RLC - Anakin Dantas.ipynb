```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

```python
L = 1000
C = 1/1000
R = 1000/1.5
w = 1/((L*C)**0.5)
V_o = 1
I_o = 1
```

```python
class Complex:                              
        def __init__(self,x,y):
            self.re = x
            self.im = y

        def soma(self, other):
            return Complex2(self.re + other.re , self.im + other.im)

        def subs(self,other):
            return Complex2(self.re - other.re , self.im - other.im)

        def mult(self,other):
            return Complex2(self.re*other.re - self.im*other.im,
                                self.re*other.im + self.im*other.re )
        def abs(self):
            return (self.re**2 + self.im**2)**0.5
        
        def theta(self):
            return(np.arctan(self.im/self.re))
        
        def mod(self):
            return((self.re**2 + self.im**2)**0.5)

        def __repr__(self):
            return '(%f, %f)' %(self.re, self.im)
```

```python
z = Complex(R, 1/(w*C) - w*L)
```

```python
t = np.linspace(0, 10, 100)
```

```python
I = Complex(V_o/z.mod()*np.cos(w*t + z.theta()), - V_o/z.mod()*np.sin(w*t + z.theta()))
```

```python
V = Complex(I_o/z.mod()*np.cos(w*t + z.theta()), - I_o/z.mod()*np.sin(w*t + z.theta()))
```

```python
fig, ax = plt.subplots()
ax.plot(t, I.re, label=r"$I$")
ax.plot(t, V.re, label=r"$V$")

ax.legend(loc=2) 
ax.set_xlabel(r'$t(s)$')
ax.set_ylabel(r'$I(A)$ $e$ $V$')
ax.set_title('Corrente e Voltagem em função do tempo');
```

```python
W = np.linspace(0, 2/((L*C)**0.5), 100)
z1 = Complex(1000/1.5, 1/(W*C) - W*L)
z2 = Complex(1000/2, 1/(W*C) - W*L)
z3 = Complex(1000/5.2, 1/(W*C) - W*L)
```

```python
a = 1/z1.mod()
b = 1/z2.mod()
c = 1/z3.mod()
```

```python
fig, ax = plt.subplots()
ax.plot(W, a, label=r"$R = 1000/1.5$")
ax.plot(W, b, label=r"$R = 1000/2$")
ax.plot(W, c, label=r"$R = 1000/5.2$")

ax.legend(loc=2) 
ax.set_xlabel(r'$\omega$')
ax.set_ylabel(r'$1/|Z|$')
ax.set_title('$1/|Z|$ vs $\omega$');
```
