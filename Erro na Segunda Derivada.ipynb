```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import math
```

```python
def f(x):
    return np.exp(x)

erro = []
def D2f(f,x):
    
    dx = x[1]-x[0]
    xplus = x+dx
    xmin = x-dx
    seg_derivada = (f(xplus)-2.*f(x) + f(xmin))/(dx*dx)
    return seg_derivada

nv = []

for i in range (1,10):
    Nvec = 2**i
    x = np.linspace(0.,10.,Nvec)
    dx = x[1]-x[0]
    a = f(x)
    nv.append(Nvec)
    erro.append((D2f(f, x)[-1] - f(x)[-1])/f(x)[-1])
```

```python
plt.loglog(nv, erro,lw=3, color='blue', label='Erro na segunda derivada')
plt.xlabel('N')
plt.ylabel(r'$\epsilon$')
plt.title("Erro no cálculo da Segunda Derivada numérica")
plt.legend()
plt.show()
```
